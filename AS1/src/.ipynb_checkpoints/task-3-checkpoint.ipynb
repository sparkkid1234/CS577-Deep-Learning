```python
import keras
import keras.backend as K
import numpy as np
import pandas as pd
from keras.utils import to_categorical
from keras import layers
from keras import models, optimizers
from keras.metrics import mse, binary_crossentropy
import os
import sys
module_path = os.path.abspath(os.path.join('./impl/'))
if module_path not in sys.path:
    sys.path.append(module_path)
from impl.DenseNeuralNet import *
from sklearn.utils import shuffle
from sklearn.model_selection import KFold
```

```python
X_train, y_train, X_test, y_test, headers, ind, impute = load_crime_data()
```

```python
X_train.shape
```

```python
removed = []
imputed = []
for i in ind:
    removed.append(headers[i])
for i in impute:
    imputed.append(headers[i])
print('Removed the following features: {}'.format(removed))
print()
print('Impute (replace by mean) the following features:{}'.format(imputed))
```

```python
hyperparams = dict(out_act=None,learning_rate=0.001,loss = "mse",metrics=["mae"])

#4-fold
k=4
kf = KFold(n_splits = k)
res = []
dnn = None
all_mae_hist = []
all_mse_hist = []
for train_ind, test_ind in kf.split(X_train):
    X_tr, X_val = X_train[train_ind], X_train[test_ind]
    y_tr, y_val = y_train[train_ind], y_train[test_ind]
    
    dnn = DNN(X_tr.shape[1],1, 3, 128,dropout=False)
    dnn.fit(X_tr, y_tr,hyperparams,epochs=500,batch_size=64,validation_data=(X_val,y_val),verbose=0)
    hist = dnn.history.history
    
    all_mae_hist.append(hist['val_mean_absolute_error'])
    all_mse_hist.append(hist['val_loss'])
    val_mse, val_mae = dnn.model.evaluate(X_val,y_val,verbose=0)
    res.append(val_mae)
```

```python
all_mae_hist = np.array(all_mae_hist)
all_mse_hist = np.array(all_mse_hist)
```

```python
avg_fold_performance = np.average(all_mae_hist,axis = 0)
avg_fold_loss = np.average(all_mse_hist,axis = 0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(avg_fold_performance)
plt.xlabel("Epochs")
plt.ylabel("Avg Validation MAE")
plt.title("Average MAE across k={} folds vs Epochs".format(k))
#plt.savefig('task3_mae_avg')
```

```python
plt.plot(avg_fold_loss)
plt.xlabel("Epochs")
plt.ylabel("Avg Validation MSE")
plt.title("Average MSE across k={} folds vs Epochs".format(k))
#plt.savefig('task3_mse_avg')
```

```python
print("Avg mae when evaluated on validation set: {}".format(np.mean(res)))
```

### Retrain

```python
hyperparams = dict(out_act=None,learning_rate=0.001,loss = "mse",metrics=["mae"])
dnn = DNN(X_train.shape[1],1, 3, 128,dropout=False)
dnn.fit(X_train, y_train,hyperparams,epochs=500,batch_size=64,verbose=0)
hist = dnn.history.history
```

```python
#Acc when dropout using higher lr
dnn.model.evaluate(X_test,y_test)
```
